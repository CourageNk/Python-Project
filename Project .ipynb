## Numerical Analysis

```python
# Describe
df.describe().T
```

```python
for i in df.describe():
    print ('Total numbers of unique values in {}'.format(i))
    print (df[i].nunique())
    print ("-" * 30)
    print ('')
```

```python
df.groupby("Charging_Habit")[['Electricity_Consumption (kWh)', 'Number_of_EVs','Transmission_Line_Capacity (MW)', 'Temperature (°C)', 'Precipitation (mm)']].max()
```

```python
def Range(series):
    return series.max()-series.min()
```

```python
df[['Number_of_EVs',"Precipitation (mm)"]].agg(Range)
```

```python
df['Precipitation (mm)'].between(.1, .8)
```

```python
for i in df.describe():
    print('Quantile value for {}'.format(i))
    print(df[i].quantile(.1, 'lower'))
```
